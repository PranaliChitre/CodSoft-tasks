```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
df = pd.read_csv("creditcard.csv")
df.head()
```

```python
df.tail()
```

```python
df.describe()
```

```python
df.info()
```

```python
# no. of missing values in each column
df.isnull()
```

```python
df['Class'].value_counts()
```

```python
# 0 == normal trans
# 1 == fraud
```

```python
legit = df[df.Class==0]
fraud = df[df.Class==1]
```

```python
legit.shape
```

```python
fraud.shape
```

```python
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
df.groupby('Class').mean()
```

```python
# Under sampling
# no. of fraudulent transaction = 492
```

```python
legit_sample = legit.sample(n = 492)
```

```python
con = pd.concat([legit_sample , fraud] , axis = 0)
con
```

```python
con['Class'].value_counts()
```

```python
con.groupby('Class').mean()
```

```python
x = con.drop(columns='Class' , axis=1)
y = con['Class']
```

```python
x
```

```python
y
```

```python
x_train , x_test , y_train , y_test = train_test_split(x, y , test_size=0.2 , stratify=y , random_state=2)
```

```python
len(x_train)
```

```python
len(x_test)
```

```python
model = LogisticRegression()
```

```python
model.fit(x_train , y_train)
```

```python
x_train_predict = model.predict(x_train)
training_acc = accuracy_score(x_train_predict , y_train)
```

```python
print('Accuracy on training data = ',training_acc)
```

```python
x_test_predict = model.predict(x_test)
testing_acc = accuracy_score(x_test_predict , y_test)
```

```python
print('Accuracy on testing data = ',testing_acc)
```

```python
from sklearn.metrics import classification_report
y_pred = model.predict(x_test)
print("Classification Report:")
print(classification_report(y_test, y_pred))
```

```python
from sklearn.metrics import confusion_matrix
print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred))
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
cm = confusion_matrix(y_test, y_pred)
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", xticklabels=['Normal', 'Fraud'], yticklabels=['Normal', 'Fraud'])
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix Heatmap')
plt.show()
```
